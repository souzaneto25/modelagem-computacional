## Funções implementadas

- `euler(f, x0, y0, h, x_end)` — RK de primeira ordem (Euler);
- `heun(f, x0, y0, h, x_end)` — RK2 variante Heun;
- `midpoint(f, x0, y0, h, x_end)` — RK2 variante Ponto Médio;
- `ralston(f, x0, y0, h, x_end)` — RK2 variante Ralston;
- `rk3(f, x0, y0, h, x_end)` — RK de terceira ordem;
- `rk4(f, x0, y0, h, x_end)` — RK de quarta ordem clássico.

Todas aceitam `f(x, y)` escalar **ou** vetorial (sistemas de EDOs).
O estado inicial `y0` pode ser escalar ou array.


```python
import numpy as np
import matplotlib.pyplot as plt

# −2x3 + 12x2 − 20x + 8.5
def f(x, y):
    return -2*x**3 + 12*x**2 - 20*x + 8.5

def runge_kutta_1st_order(f, x0, y0, h, x_final):

    x_values = [x0]
    y_values = [y0]

    x = x0
    y = y0

    while x < x_final:
        x1 = f
        y = y + h * f(x, y)
        x = x + h

        x_values.append(x)
        y_values.append(y)

    return np.array(x_values), np.array(y_values)

def runge_kutta_2nd_order(f, x0, y0, h, x_final):

    x_values = [x0]
    y_values = [y0]

    x = x0
    y = y0

    while x < x_final:
        k1 = f(x, y)
        k2 = f(x + h/2, y + h/2 * k1)

        y = y + h * k2
        x = x + h

        x_values.append(x)
        y_values.append(y)

    return np.array(x_values), np.array(y_values)

def runge_kutta_3rd_order(f, x0, y0, h, x_final):

    x_values = [x0]
    y_values = [y0]

    x = x0
    y = y0

    while x < x_final:
        k1 = f(x, y)
        k2 = f(x + h/2, y + h/2 * k1)
        k3 = f(x + h, y - h * k1 + 2 * h * k2)

        y = y + (h/6) * (k1 + 4*k2 + k3)
        x = x + h

        x_values.append(x)
        y_values.append(y)

    return np.array(x_values), np.array(y_values)

def runge_kutta_4th_order(f, x0, y0, h, x_final):

    x_values = [x0]
    y_values = [y0]

    x = x0
    y = y0

    while x < x_final:
        k1 = f(x, y)
        k2 = f(x + h/2, y + h/2 * k1)
        k3 = f(x + h/2, y + h/2 * k2)
        k4 = f(x + h, y + h * k3)

        y = y + (h/6) * (k1 + 2*k2 + 2*k3 + k4)
        x = x + h

        x_values.append(x)
        y_values.append(y)

    return np.array(x_values), np.array(y_values)

def solucao_exata(x):
    return -0.5*x**4 + 4*x**3 - 10*x**2 + 8.5*x + 1


#Parametros do problema
x0 = 0.0
y0 = 1.0
x_final = 5.0
h=0.1

#Resolvendo
x_rk1, y_rk1 = runge_kutta_1st_order(f, x0, y0, h, x_final)
x_rk2, y_rk2 = runge_kutta_2nd_order(f, x0, y0, h, x_final)
x_rk3, y_rk3 = runge_kutta_3rd_order(f, x0, y0, h, x_final)
x_rk4, y_rk4 = runge_kutta_4th_order(f, x0, y0, h, x_final)


#Solucao exata
x_exact = np.linspace(x0, x_final, 5000)
y_exact = solucao_exata(x_exact)

#Tabela de resultados
print(f"{'x':>10} {'RK1':>10} {'RK2':>10} {'RK3':>10} {'RK4':>10} {'Exata':>10}")
for i in range(len(x_rk1)):
    print(f"{x_rk1[i]:10.4f} {y_rk1[i]:10.4f} {y_rk2[i]:10.4f} {y_rk3[i]:10.4f} {y_rk4[i]:10.4f} {solucao_exata(x_rk1[i]):10.4f}")


```
