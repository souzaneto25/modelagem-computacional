# Exemplo 1 — RK de Primeira Ordem (Método de Euler)

**Slide p. 11–13** — Integrar numericamente

$$\frac{dy}{dx} = -2x^3 + 12x^2 - 20x + 8{,}5$$

com $x \in [0, 1]$, $h = 0{,}5$ e $y(0) = 1$.

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x, y):
    return -2*x**3 + 12*x**2 - 20*x + 8.5

def y_exato(x):
    return -0.5*x**4 + 4*x**3 - 10*x**2 + 8.5*x + 1

x0, y0, h = 0.0, 1.0, 0.5
x_end = 1.0

x, y = x0, y0
print(f"x = {x:.1f}   y_euler = {y:.5f}   y_exato = {y_exato(x):.5f}")
while x < x_end - 1e-10:
    y = y + h * f(x, y)
    x = x + h
    print(f"x = {x:.1f}   y_euler = {y:.5f}   y_exato = {y_exato(x):.5f}   erro = {abs(y - y_exato(x)):.5f}")
```
