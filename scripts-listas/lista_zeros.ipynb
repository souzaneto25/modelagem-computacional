## Questão 1
f(s) = s³+7s²+16s+12 = (s+2)²(s+3). Raiz simples s=−3 e raiz dupla s=−2.

```python
def f(s): return s**3 + 7*s**2 + 16*s + 12
def df(s): return 3*s**2 + 14*s + 16

print("Raiz simples s=-3 (secante modificada)")
for delta in [0.01, 0.05]:
    x = -3.5
    for k in range(100):
        dx = delta*abs(x) if x != 0 else delta
        x = x - f(x)*dx / (f(x+dx) - f(x))
        if abs(f(x)) < 1e-6: break
    print(f"  delta={delta}: s ≈ {x:.8f}  ({k+1} iter)")

print("\nRaiz dupla s=-2 (Newton padrão)")
x = -1.5
for k in range(100):
    xn = x - f(x)/df(x)
    if abs(xn-x) < 1e-6 or abs(f(xn)) < 1e-6: x = xn; break
    x = xn
print(f"  s ≈ {x:.8f}  ({k+1} iter)")

print("\nRaiz dupla s=-2 (Newton modificado m=2)")
x = -1.5
for k in range(100):
    xn = x - 2*f(x)/df(x)
    if abs(xn-x) < 1e-6: x = xn; break
    x = xn
print(f"  s ≈ {x:.8f}  ({k+1} iter)")
```

## Questão 2
Sistema: x²+y²=16 e (x−4)²+(y−4)²=5. Newton 2×2.

```python
for x0, y0 in [(3.8, 1.2), (1.2, 3.8)]:
    x, y = x0, y0
    for k in range(100):
        F1 = x**2 + y**2 - 16
        F2 = (x-4)**2 + (y-4)**2 - 5
        J = np.array([[2*x, 2*y],[2*(x-4), 2*(y-4)]])
        dx, dy = np.linalg.solve(J, [-F1, -F2])
        x += dx; y += dy
        if abs(dx) < 1e-8 and abs(dy) < 1e-8: break
    print(f"x0={x0}: ({x:.6f}, {y:.6f})  ({k+1} iter)")
```

## Questão 3
Ângulo de disparo α: (220√2/π)(1+cosα) = 100.

```python
import numpy as np
Tmax = 220*np.sqrt(2)
f  = lambda a: Tmax/np.pi*(1+np.cos(a)) - 100
df = lambda a: -Tmax/np.pi*np.sin(a)
x = 1.0
for k in range(100):
    xn = x - f(x)/df(x)
    if abs(xn-x) < 1e-8: x = xn; break
    x = xn
print(f"α = {x:.8f} rad = {np.degrees(x):.6f}°  ({k+1} iter)")
```

## Questão 4
Indutância L: frequência de ressonância fc = 1000 Hz, C = 1 µF.

```python
C, fc = 1e-6, 1000.0
f  = lambda L: 1/(2*np.pi*np.sqrt(L*C)) - fc
df = lambda L: -1/(4*np.pi*np.sqrt(C)*L**1.5)
x = 0.01
for k in range(100):
    xn = x - f(x)/df(x)
    if abs(xn-x) < 1e-10: x = xn; break
    x = xn
print(f"L = {x:.10f} H  ({k+1} iter)")
print(f"Conferência: {1/((2*np.pi*fc)**2*C):.10f} H")
```

## Questão 5
Frequência de ressonância f₀ em circuito RLC (R=10 Ω, L=1 mH, C=1 µF).

```python
R, L, C = 10.0, 1e-3, 1e-6
h  = lambda f: 2*np.pi*f*L - 1/(2*np.pi*f*C)
dh = lambda f: 2*np.pi*L + 1/(2*np.pi*C*f**2)
x = 5000.0
for k in range(100):
    xn = x - h(x)/dh(x)
    if abs(xn-x) < 1e-6: x = xn; break
    x = xn
print(f"f₀ = {x:.6f} Hz  ({k+1} iter)")
print(f"Ideal 1/(2π√LC) = {1/(2*np.pi*np.sqrt(L*C)):.6f} Hz")
```

## Questão 6
Frequência F em circuito RLC (R=140 Ω, L=260 mH, C=25 µF, vm=24 V, im=0.15 A).

```python
R, L, C, vm, im = 140.0, 260e-3, 25e-6, 24.0, 0.15
f  = lambda F: vm/np.sqrt(R**2 + (2*np.pi*F*L - 1/(2*np.pi*F*C))**2) - im
df = lambda F: (-vm*(2*(2*np.pi*F*L-1/(2*np.pi*F*C))*(2*np.pi*L+1/(2*np.pi*C*F**2)))
                / (2*(R**2+(2*np.pi*F*L-1/(2*np.pi*F*C))**2)**1.5))
x = 30.0
for k in range(100):
    xn = x - f(x)/df(x)
    if abs(xn-x) < 1e-6: x = xn; break
    x = xn
print(f"F = {x:.6f} Hz  ({k+1} iter)")
```

## Questão 7
Tensão de máxima potência Vmp de célula solar (Voc = 0.5 V, T = 297 K).

```python
q, kB, Voc, T = 1.6022e-19, 1.3806e-23, 0.5, 297.0
f  = lambda V: np.exp(q*V/(kB*T))*(1 + q*V/(kB*T)) - np.exp(q*Voc/(kB*T))
df = lambda V: np.exp(q*V/(kB*T))*(q/(kB*T))*(2 + q*V/(kB*T))
x = 0.4
for k in range(100):
    xn = x - f(x)/df(x)
    if abs(xn-x) < 1e-10: x = xn; break
    x = xn
print(f"Vmp = {x:.8f} V  ({k+1} iter)")
```
